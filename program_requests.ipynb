```python
import requests
import json
from datetime import datetime, timedelta
import googlemaps
import geocoder

# timestamp = 1703019206145 / 1000  # Assuming the timestamp is in milliseconds, divide by 1000 to convert to seconds

# # Convert the timestamp to a datetime object
# dt_object = datetime.utcfromtimestamp(timestamp)

# # Format the datetime object as a string
# formatted_datetime = dt_object.strftime('%Y-%m-%d %H:%M:%S UTC')

# print("Converted DateTime:", formatted_datetime)


url = "https://store.transitstat.us/passio_go/rutgers"  # Replace this with the actual URL of the website

# Make an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data from the response
    data = response.json()

    # You can iterate through the data to access other information as needed
    for train_id, train_info in data["trains"].items():
        print(f"Bus {train_id}, {train_info['line']}: Destination - {train_info['dest']}")
        for prediction in train_info["predictions"]:
            timestamp = prediction["actualETA"] / 1000
            dt_object = datetime.utcfromtimestamp(timestamp)
            est_offset = -5  # EST is UTC-5
            est_timedelta = timedelta(hours=est_offset)
            dt_est = dt_object + est_timedelta
            
            formatted_datetime = dt_est.strftime('%H:%M:%S')
            print(f"Station {prediction['stationName']}: ETA - {formatted_datetime}")
        print()
else:
    print(f"Error: Unable to retrieve data. Status code: {response.status_code}")
```
