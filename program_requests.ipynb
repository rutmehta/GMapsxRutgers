```python
import requests
from datetime import datetime, timedelta
import googlemaps
import pandas as pd
import json

# timestamp = 1703019206145 / 1000  # Assuming the timestamp is in milliseconds, divide by 1000 to convert to seconds

# # Convert the timestamp to a datetime object
# dt_object = datetime.utcfromtimestamp(timestamp)

# # Format the datetime object as a string
# formatted_datetime = dt_object.strftime('%Y-%m-%d %H:%M:%S UTC')

# print("Converted DateTime:", formatted_datetime)


url = "https://store.transitstat.us/passio_go/rutgers"  # Replace this with the actual URL of the website

# Make an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Assuming `response` is the response object from a web request
    columns = ['bus_id', 'line', 'lineCode', 'dest', 'stationID', 'stationName', 'actualETA', 'noETA']
    
    data = response.json()['trains']
    print(data)
    
    
    results = pd.DataFrame()
    
    for bus_id, bus_data in data.items():
        for predictions in bus_data['predictions']:
            temp = pd.DataFrame([bus_id, bus_data['line'], bus_data['lineCode'], bus_data['dest'], predictions['stationID'], predictions['stationName'], predictions['actualETA'], predictions['noETA']]).T
            temp.columns = columns
            results = pd.concat([results, temp], ignore_index=True)
    display(results)
else:
    print(f"Error: Unable to retrieve data. Status code: {response.status_code}")
    
origin = "Livingston Plaza"
destination1 = "Science Building"
destination2 = "Allison Road Classrooms"

filter = results[(results['stationName'] == origin) | (results['stationName'] == destination1) | (results['stationName'] == destination2)]
display(filter)
```
